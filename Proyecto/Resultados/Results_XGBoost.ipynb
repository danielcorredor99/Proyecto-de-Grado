```python
def makePlot(lags_label,lags,neurons,attributes):
    #mse data
    train_plot_data_mse={lags_label[i]:{j:lags[i][j]["train"]["MSE"] for j in neurons} for i in range(attributes)}
    test_plot_data_mse={lags_label[i]:{j:lags[i][j]["test"]["MSE"] for j in neurons} for i in range(attributes)}
    test_df_mse=pd.DataFrame(test_plot_data_mse)
    train_df_mse=pd.DataFrame(train_plot_data_mse)

    #mae
    train_plot_data_mae={lags_label[i]:{j:lags[i][j]["train"]["MAE"] for j in neurons} for i in range(attributes)}
    test_plot_data_mae={lags_label[i]:{j:lags[i][j]["test"]["MAE"] for j in neurons} for i in range(attributes)}
    test_df_mae=pd.DataFrame(test_plot_data_mae)
    train_df_mae=pd.DataFrame(train_plot_data_mae)

    #rmse data
    train_plot_data_rmse={lags_label[i]:{j:lags[i][str(j)]["train"]["RMSE"] for j in neurons} for i in range(attributes)}
    test_plot_data_rmse={lags_label[i]:{j:lags[i][str(j)]["test"]["RMSE"] for j in neurons} for i in range(attributes)}
    test_df_rmse=pd.DataFrame(test_plot_data_rmse)
    train_df_rmse=pd.DataFrame(train_plot_data_rmse)

    #mape
    train_plot_data_mape={lags_label[i]:{j:lags[i][j]["train"]["MAPE"] for j in neurons} for i in range(attributes)}
    test_plot_data_mape={lags_label[i]:{j:lags[i][j]["test"]["MAPE"] for j in neurons} for i in range(attributes)}
    test_df_mape=pd.DataFrame(test_plot_data_mape)
    train_df_mape=pd.DataFrame(train_plot_data_mape)
    
    ##Plot ---------------
    fig, ax = plt.subplots(4,2,figsize=[15,25])
    ax[0][0].plot(train_df_mse,label=lags_label)
    ax[1][0].plot(train_df_mae,label=lags_label)
    ax[2][0].plot(train_df_rmse,label=lags_label)
    ax[3][0].plot(train_df_mape,label=lags_label)
    ax[0][1].plot(test_df_mse,label=lags_label)
    ax[1][1].plot(test_df_mae,label=lags_label)
    ax[2][1].plot(test_df_rmse,label=lags_label)
    ax[3][1].plot(test_df_mape,label=lags_label)
    ax[0][0].set(xlabel="neurons",title="Train MSE")
    ax[1][0].set(xlabel="neurons",title="Train MAE")
    ax[2][0].set(xlabel="neurons",title="Train RMSE")
    ax[3][0].set(xlabel="neurons",title="Train MAPE")

    ax[0][1].set(xlabel="neurons",title="Test MSE")
    ax[1][1].set(xlabel="neurons",title="Test MAE")
    ax[2][1].set(xlabel="neurons",title="Test RMSE")
    ax[3][1].set(xlabel="neurons",title="Test MAPE")

    ax[0][0].legend(loc="upper right")
    ax[1][0].legend(loc="upper right")
    ax[1][1].legend(loc="upper right")
    ax[0][1].legend(loc="upper right")

    ax[2][0].legend(loc="upper right")
    ax[2][0].legend(loc="upper right")
    ax[3][1].legend(loc="upper right")
    ax[3][1].legend(loc="upper right")
```
