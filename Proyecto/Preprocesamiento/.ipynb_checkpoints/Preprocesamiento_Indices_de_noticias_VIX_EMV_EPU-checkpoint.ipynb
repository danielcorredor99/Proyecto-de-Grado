```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import NullFormatter
from datetime import datetime
from seaborn import boxplot
from seaborn import distplot
from tsmoothie.utils_func import sim_randomwalk, sim_seasonal_data
from tsmoothie.smoother import *
from statsmodels.tsa.stattools import adfuller
sns.set()
```

## Columna de Fecha

```python
df_fecha = pd.date_range(name= 'Fecha', start='2010-01-01', end='2022-07-06', freq='D')
df_fecha = pd.DataFrame(df_fecha, dtype= 'datetime64[ns]')
df_fecha = df_fecha.sort_values(by='Fecha',ascending=True, ignore_index=True)
```

## Importar dataframe VIX

```python
df_vix = pd.read_csv(r"Datos en bruto/VIX_TODO.csv")
df_vix['Fecha']  = pd.to_datetime(df_vix['Fecha'])   
df_vix = df_vix.set_index('Fecha')
df_vix = df_fecha.set_index('Fecha').combine_first(df_vix)
df_vix.head()
```

## Importar dataframe EMV-ID

```python
df_emv = pd.read_csv(r"Datos en bruto\IDEMV_TODO.csv")
df_emv = df_emv.query("Fecha > '2009-12-31' and Fecha <= '2022-07-06'")
df_emv['Fecha'] = pd.to_datetime(df_emv['Fecha'])   
df_emv.set_index('Fecha', inplace=True)
df_emv.head()
```

## Importar dataframe EPU

```python
df_epu = pd.read_csv(r"Datos en bruto\EPU_TODO.csv")
df_epu = df_epu.query("Fecha > '2009-12-31' and Fecha <= '2022-07-06'")
df_epu['Fecha'] = pd.to_datetime(df_epu['Fecha'])   
df_epu.set_index('Fecha', inplace=True)
df_epu.head()
```

## ImputaciÃ³n de datos

```python
df_vix =  df_vix.interpolate(method='cubicspline', order=3).query("Fecha > '2009-12-31' and Fecha <= '2022-07-06'")
df_vix.count()
```

```python
df_ind_noticias = pd.concat([df_vix, df_emv , df_epu], axis=1)
df_ind_noticias
```

## Prueba de estacionariedad - MÃ©todo: Pureba Dickey Fuller Aumentada

```python
### Prueba de estacionalidad - VIX

adft = adfuller(df_ind_noticias['VIX'],autolag="t-stat")
output_df = pd.DataFrame({"Metric":["Test Statistics","p-value","No. of lags used","Number of observations used", 
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"], "Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]})
print('Prueba Dickey Fuller aumentada - VIX')
print(output_df); 
print("")

if adft[1] > 0.05:
    print("La serie de tiempo del VIX No es estacionaria")
else:
    print("La serie de tiempo del VIX es estacionaria");
```

```python
### Prueba de estacionalidad - EMV-ID

adft = adfuller(df_ind_noticias['EMV-ID'],autolag="t-stat")
output_df = pd.DataFrame({"Metric":["Test Statistics","p-value","No. of lags used","Number of observations used", 
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"], "Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]})
print('Prueba Dickey Fuller aumentada - EMV')
print(output_df); 
print("")

if adft[1] > 0.05:
    print("La serie de tiempo del EMV No es estacionaria")
else:
    print("La serie de tiempo del EMV es estacionaria");
```

```python
### Prueba de estacionalidad - EPU

adft = adfuller(df_ind_noticias['EPU'],autolag="t-stat")
output_df = pd.DataFrame({"Metric":["Test Statistics","p-value","No. of lags used","Number of observations used", 
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"], "Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]})
print('Prueba Dickey Fuller aumentada - EPU')
print(output_df); 
print("")

if adft[1] > 0.05:
    print("La serie de tiempo del EPU No es estacionaria")
else:
    print("La serie de tiempo del EPU es estacionaria")
```
