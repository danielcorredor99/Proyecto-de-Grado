```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import NullFormatter
from datetime import datetime
from seaborn import boxplot
from seaborn import distplot
sns.set()
```

```python
df = pd.read_csv("C:/Users/Lenovo/Documents/2. Proyecto de grado\Datos históricos Futuros café COL.csv")
df0 = df[["Fecha", "Último"]]
df0['Fecha']=pd.to_datetime(df['Fecha'], format="%d.%m.%Y")
df0['Último'] = df0['Último'].apply(lambda x: x.replace(',','.'))
df0['Último'] = df0['Último'].astype(float)
```

```python
df0.describe()
```

```python
fig, ax = plt.subplots(figsize = (7, 7))
sns.boxplot(data = df0[['Último']])
```

```python
df0['Último'].hist(figsize=(7,5))
```

```python
fig, ax = plt.subplots(figsize = (20,5))
sns.lineplot(x = "Fecha", y = "Último", data = df0)
plt.xlabel("Fecha") 
plt.ylabel("Precio futuro"); 
```
